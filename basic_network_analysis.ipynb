```python
import pandas as pd
import networkx as nx
import operator
import numpy as np
```

```python
G = nx.Graph()
```

```python
G.add_node("친구")
```

```python
G.add_nodes_from(['친구','가족','직장동료'])
```

```python
G.nodes()
```

```python
G.add_edge('나','가족')
```

```python
G.add_edges_from([('나','가족'),('나','친구'),('나','직장동료')])
```

```python
G.edges()
```

네크워크 크기(노드수)

```python
G = nx.Graph()
relations = [('나', '친구'), ('나', '가족'), ('나', '직장동료')]
G.add_edges_from(relations)
len(G)
G.order()
```

```python
G = nx.Graph()
relations = [('나', '친구'), ('나', '가족'), ('나', '직장동료')]
G.add_edges_from(relations)
nx.classes.function.density(G)
```
