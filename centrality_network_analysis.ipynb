```python
import pandas as pd
import networkx as nx
import operator
import numpy as np
```

연결정도 중심성(Degree Centrality)

```python
g = nx.Graph()
```

```python
relations = [('A', 'B'), ('A', 'C'), ('A', 'D'), ('A', 'E'), ('A', 'F'), ('A', 'G'), ('B', 'C'), ('B', 'D'), ('B', 'E'), ('B', 'E'), ('C', 'D'), ('D', 'F'), ('E', 'F'), ('E', 'G'), ('F', 'G'), ('G', 'H'), ('H', 'I'), ('H', 'K'), ('I', 'J'), ('I', 'J'),('I','L'),('J','L')]
```

```python
g.add_edges_from(relations)
```

```python
#네트워크의 크기(노드수)
len(g)
```

```python
#네트워크의 크기(노드수)
g.order()
```

```python
nx.draw(g,with_labels=True)
```

```python
#중심성 분석 : 노드들의 영량력 크기를 분석
#하위집단 분석 : 하위집단들이 얼마나 있나
#에고 네트워크 분석

#Degree
#연결정도 중심성 : 네트워크 내 노드들이 얼많아 많은 연결을 가지고 있는지 측정하는 지표
#임의의 한 노드에 연결이 많을 수록 해당 노드는 더 많은 자율성과 권력을 가진다는 개념

#Closeness
#근접중심성 : 네트워크 내의 임의의 한 노드가 다른 모든 노드들과 얼마나 근접하게 연결되어 있는지
#해당 노드가 얼마나 네트워크의 중앙에 위치하는지 측정하는 지표

#매개중신성
#임의의 한 노드가 다른 노드가 네트워크를 구축할 때 중개자 역할을 얼마나 많이 수행하는지 측정하는 지표
#정보의 흐름을 통제할 수 있는 큰 영향력을 가지게 됨(포털 사이트)

#위세중심성
#영향력, 중요도
```

```python
#네트워크 밀도(무방향) 계산
nx.density(g)
```

```python
#상대적 연결정도 중심성(무방향) 계산
nx.degree_centrality(g)
#A 가 가장 높다 A를 중심으로 촘촘하게 연결되어 있다.
```

근접중심성

```python
#상대적 근접중심성 계산
nx.closeness_centrality(g)
#G 가 가장 높다 G가 가장 중앙에 위치해 있다.
```

```python
#노드G의 상대적 근접 중심성 계산
nx.closeness_centrality(g,"G")
```

매개중심성

```python
nx.betweenness_centrality(g)
#H 가 가장 매개적인 위치에 있다. 
```

아이겐벡터 중심성

```python
nx.eigenvector_centrality(g)
#세력의 정도
```
